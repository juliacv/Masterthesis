## Exploratory data analysis
This should be used with the data from the Final_data file. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Original Data
y = np.array([1632, 1705, 3500, 1002, 344, 6596])
labels = ["Anastomosis (1632)", "Aorta (1705)", "Vessel (3500)", "Questionable (1002)", "Blank (344)", "Unknown (6596)"]

fig, ax = plt.subplots()
ax.pie(y, labels=labels, autopct='%1.0f%%', colors = ['peachpuff','pink','lightsteelblue', 'lavender','lightgray', 'lemonchiffon'])
plt.title("Annotations of REQUEST data")
```

```python
#Division of the data in the training set before data augmentation
y_2 = np.array([1160, 1192, 2432, 249, 4588])
labels_2 = [f"Anastomosis ({1160})", f"Aorta ({1192})", f"Vessel ({2432})", "Blank (249)", "Unknown (4588)"]

fig, ax = plt.subplots()
ax.pie(y_2, labels=labels_2, autopct='%1.0f%%', colors = ['peachpuff','pink','lightsteelblue','lightgray', 'lemonchiffon'])
plt.title("Annotations of REQUEST data")
```

```python
#Division of data after data augmentation
y_2 = np.array([5103, 5241, 10744, 344, 6596])
labels_2 = [f"Anastomosis ({5103})", f"Aorta ({5241})", f"Vessel ({10744})", "Blank (344)", "Unknown (6596)"]

fig, ax = plt.subplots()
ax.pie(y_2, labels=labels_2, autopct='%1.0f%%', colors = ['peachpuff','pink','lightsteelblue','lightgray', 'lemonchiffon'])
plt.title("Annotations of REQUEST data")
```

```python
#Division of training data after data augmentation
y_2 = np.array([4634, 4741, 9692, 249, 4588])
labels_2 = [f"Anastomosis ({4634})", f"Aorta ({4741})", f"Vessel ({9692})", "Blank (249)", "Unknown (4588)"]

fig, ax = plt.subplots()
ax.pie(y_2, labels=labels_2, autopct='%1.0f%%', colors = ['peachpuff','pink','lightsteelblue','lightgray', 'lemonchiffon'])
plt.title("Annotations of REQUEST data")
```

```python
#Number of Vessels after Data Augmentation
Vessels = ("1 vessel", "2 vessels", "3 vessels", "4 vessels", "5 vessels")
y_pos = np.arange(len(Vessels))
Number = [8500, 1670, 495, 75, 4]

fig, ax = plt.subplots()
plt.ylabel("Number of images")
plt.title("Number of vessels per image")
plt.xticks(y_pos, Vessels)

bars = ax.bar(y_pos, Number, align="center", alpha=0.5)

for b in bars:
    height = b.get_height()
    ax.text(x=b.get_x() + b.get_width() / 2, y=height+.10, s="{}".format(height), ha='center', va="bottom")

plt.show()


```
